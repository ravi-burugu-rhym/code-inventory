```python
# function to extract text from all pages
def raw_pdf(x):
    page=[]
    for i in  range(0,x.numPages): 
        pages=x.getPage(i) 
        story=pages.extractText() 
        k= ' '.join (story.split())
        page.append(k)
    return page
```

```python
# removing special symbols
def remove_spl(data):
    new=[]
    for i in range(test.numPages):
        p=re.sub('[ﬁ,ﬂ]',' ',data[i])
        new.append(p)
    return(new)
```

```python
# extraxting data
data=raw_pdf(test)
```

```python
clean_data=remove_spl(data)
```

```python
# removing p
pun=list(punctuation)
k=[w for w in pun if (w!='.' ) & (w!='(') & (w!=')')]
```

```python
def text_processing(text):
    pre_processed_text=[]
    words = [w for sent in nltk.sent_tokenize(str(text)) for w in nltk.word_tokenize(sent)]    
    stop_words=stopwords.words("english")+k
    tokens=[w for w in words if w not in stop_words] 
    tokens=[word for word in tokens]
    pre_processed_text.append(" ".join(tokens))
    return pre_processed_text
```
