```python
#importing of required libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
def hashtag_words(start,end,country,lang,keys=[],*args):
        tags=[]
        temp=''
        for i in range(len(keys)):
            
            if not keys[i].startswith('#'):
                if i==0:
                    word= str(keys[i])
                    tags.append(word)
                if i>0:
                    word= str(keys[i])
                    tags.append('%2C'+word)
            elif '#' in keys[i]:
                word=keys[i].replace('#','%23')
                tags.append(word)
        k=temp.join(tags)
        
        final='https://twitter.com/search?f=tweets&vertical=news&q='+k
        
        if country is not None:
            country='%20near%3A%22'+country+'%22%20within%3A1000mi'
            final=final+country

            
        if start is not None:
            start='%20since%3A'+start
            final=final+start
        
        if end is not None:
            end='%20until%3A'+end
            final=final+end
        
        if  lang is not None:
            lang='&l='+lang
            final=final+lang
       
            
        final_link=final+'&src=typd'
            
            
        #final_link='https://twitter.com/search?f=tweets&vertical=news&q='+country+start+end+lang+'&src=typd'
       
        return final_link
```

```python
def twitter_posts(hashtags,start,end,lang,csv_filepath,country):
    
    driver = webdriver.Chrome(r"C:\Users\admin\Desktop\RHYMcygiene\database\chromedriver.exe")
    k=hashtag_words(keys=hashtags,country=country,start=start,end=end,lang=lang,)
    print(k)
    driver.maximize_window() # to maximize window

    driver.get(k)
    SCROLL_PAUSE_TIME = 1.9

    
    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        
    
    content = driver.page_source
    soup = BeautifulSoup(content)
    
    main_link=soup.find_all('li',class_="js-stream-item stream-item stream-item")
    
    
    twitter_id=[]
    reference_link=[]
    twitter_post=[]
    link_post=[]
    
    for i in range(len(main_link)):
        
        name=main_link[i].find('strong').text
        
        profile=main_link[i].find('a',class_='account-group js-account-group js-action-profile js-user-profile-link js-nav').get('href')
        
        profile_link='https://twitter.com'+profile
        #print(profile_link)
        
        nick=main_link[i].find('span',class_="username u-dir u-textTruncate").text.replace('@','')
        ID=main_link[i].get('data-item-id')
        post_link='https://twitter.com/'+nick+'/status/'+ID
     
        #link=main_link[i].find('div',class_="tweet js-stream-tweet js-actionable-tweet js-profile-popup-actionable dismissible-content original-tweet js-original-tweet has-cards has-content ")
        #if link is None:
            #link=main_link[i].find('div',class_="tweet js-stream-tweet js-actionable-tweet js-profile-popup-actionable dismissible-content original-tweet js-original-tweet ").get('data-permalink-path')
            #post_link='https://twitter.com'+link
        #else:
            #link=main_link[i].find('div',class_="tweet js-stream-tweet js-actionable-tweet js-profile-popup-actionable dismissible-content original-tweet js-original-tweet has-cards has-content ").get('data-permalink-path')
            #post_link='https://twitter.com'+link
        #print(post_link)
        
        post=main_link[i].find('p').text.replace('\n','')   
        
        
        
        twitter_id.append(name)
        reference_link.append(profile_link)
        twitter_post.append(post)
        link_post.append(post_link)
        
        
    main_data={'Name':twitter_id,
                'ref_links':reference_link,
                'post':twitter_post,
                'linkedin_post_link': link_post}
        
    post_df=pd.DataFrame(main_data)
    new_df=post_df.drop_duplicates(subset ="post", keep = False) 
    
    # created post_dataframe is saved to csv file 
    new_df.to_csv(csv_filepath)
        
    return post_df
        
        
        
    
    
        
```
